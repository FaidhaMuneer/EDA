```python
spark = SparkSession.builder \
    .appName("Medical Fraud Detection EDA") \
    .getOrCreate()
```

```python
file_path = "/content/drive/MyDrive/Colab Notebooks/big data/dftocsv.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)
```

```python
df.createOrReplaceTempView("df")
```

```python
import matplotlib.pyplot as plt

query_result = spark.sql("""
    SELECT IS_FRAUD, total_Tot_Drug_Cst
    FROM df
""")

plt.figure(figsize=(10, 6))
query_result.toPandas().boxplot(by="IS_FRAUD", figsize=(10, 6))
plt.title("Box Plot of Total Drug Cost by Fraud")
plt.xlabel("Fraud")
plt.ylabel("Total Drug Cost")
plt.show()
```

```python
query_result = spark.sql("""
    SELECT total_Tot_Drug_Cst, IS_FRAUD
    FROM df
""")

plt.figure(figsize=(10, 6))
query_result.toPandas().hist(by="IS_FRAUD", bins=20, figsize=(10, 6))
plt.title("Histogram of Total Drug Cost by Fraud")
plt.xlabel("Total Drug Cost")
plt.ylabel("Frequency")
plt.show()
```

```python
query_result = spark.sql("""
    SELECT Prscrbr_Type,
           AVG(CASE WHEN IS_FRAUD = 0 THEN total_Tot_Drug_Cst END) AS non_fraud_mean,
           AVG(CASE WHEN IS_FRAUD = 1 THEN total_Tot_Drug_Cst END) AS fraud_mean
    FROM df
    GROUP BY Prscrbr_Type
    ORDER BY Prscrbr_Type
""")

plt.figure(figsize=(30, 6))
query_result.toPandas().plot(kind="bar", x="Prscrbr_Type", figsize=(30, 6))
plt.title("Mean Total Drug Cost by Prscrbr_Type and Fraud")
plt.xlabel("Prscrbr_Type")
plt.ylabel("Mean Total Drug Cost")
plt.show()
```

```python
query = """
    SELECT Prscrbr_Type, SUM(IS_FRAUD) AS total_fraud
    FROM df
    GROUP BY Prscrbr_Type
"""
query_result = spark.sql(query)

result_df = query_result.toPandas()

plt.figure(figsize=(10, 6))
plt.pie(result_df["total_fraud"], labels=result_df["Prscrbr_Type"], autopct='%1.1f%%')
plt.title("Fraud Distribution by Prscrbr_Type")
plt.axis("equal")
plt.show()
```

```python
query = """
    SELECT total_Tot_Drug_Cst, total_Tot_Clms
    FROM df
"""
query_result = spark.sql(query)

result_df = query_result.toPandas()

plt.figure(figsize=(10, 6))
plt.scatter(result_df["total_Tot_Drug_Cst"], result_df["total_Tot_Clms"])
plt.title("Scatter Plot of Total Drug Cost vs. Total Claims")
plt.xlabel("Total Drug Cost")
plt.ylabel("Total Claims")
plt.show()
```

```python
query = """
    SELECT total_Tot_Day_Suply, total_Tot_Drug_Cst
    FROM df
    ORDER BY total_Tot_Day_Suply
"""

query_result = spark.sql(query)
result_df = query_result.toPandas()

plt.figure(figsize=(10, 6))
result_df.plot(kind="area", x="total_Tot_Day_Suply", y="total_Tot_Drug_Cst", figsize=(10, 6))
plt.title("Area Plot of Total Drug Cost over Time")
plt.xlabel("Total Day Supply")
plt.ylabel("Total Drug Cost")
plt.show()

```

```python
query = """
    SELECT Prscrbr_Last_Org_Name, AVG(total_Tot_Drug_Cst) AS avg_total_Tot_Drug_Cst
    FROM df
    GROUP BY Prscrbr_Last_Org_Name
    ORDER BY avg_total_Tot_Drug_Cst
    LIMIT 10
"""
query_result = spark.sql(query)
result_df = query_result.toPandas()


plt.figure(figsize=(10, 6))
plt.plot(result_df["Prscrbr_Last_Org_Name"], result_df["avg_total_Tot_Drug_Cst"])
plt.title("Line Plot of Mean Total Drug Cost by Prscrbr_Last_Org_Name")
plt.xlabel("Prscrbr_Last_Org_Name")
plt.ylabel("Mean Total Drug Cost")
plt.xticks(rotation=45)
plt.show()

```

```python
query_fraud = spark.sql("""
    SELECT total_Tot_Drug_Cst
    FROM df
    WHERE IS_FRAUD = 1
""")
query_non_fraud = spark.sql("""
    SELECT total_Tot_Drug_Cst
    FROM df
    WHERE IS_FRAUD = 0
""")

fraud_data = query_fraud.toPandas()["total_Tot_Drug_Cst"]
non_fraud_data = query_non_fraud.toPandas()["total_Tot_Drug_Cst"]

plt.figure(figsize=(10, 6))
sns.kdeplot(fraud_data, label="Fraud", shade=True)
sns.kdeplot(non_fraud_data, label="Non-Fraud", shade=True)
plt.title("Kernel Density Estimate of Total Drug Cost by Fraud (Using SQL)")
plt.xlabel("Total Drug Cost")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
query = """
    SELECT IS_FRAUD,
           AVG(avg_Tot_Drug_Cst) AS avg_Tot_Drug_Cst,
           AVG(avg_Tot_Clms) AS avg_Tot_Clms,
           AVG(avg_Tot_Day_Suply) AS avg_Tot_Day_Suply,
           AVG(Total_Payment_Sum) AS Total_Payment_Sum
    FROM df
    GROUP BY IS_FRAUD
"""
result_df = spark.sql(query)

mean_values_pd = result_df.toPandas()

categories = list(mean_values_pd.columns[1:])
values_fraud = mean_values_pd.iloc[1, 1:].values
values_non_fraud = mean_values_pd.iloc[0, 1:].values

import numpy as np
import matplotlib.pyplot as plt

num_vars = len(categories)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

values_fraud = np.concatenate((values_fraud,[values_fraud[0]]))
values_non_fraud = np.concatenate((values_non_fraud,[values_non_fraud[0]]))
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))
ax.fill(angles, values_fraud, color='red', alpha=0.25, label='Fraud')
ax.fill(angles, values_non_fraud, color='blue', alpha=0.25, label='Non-Fraud')
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Radar Chart of Mean Feature Values by Fraud Status')
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
